```python
import requests
import xmltodict
from suds.client import Client
import pandas as pd

phone_df = pd.read_csv('Phone.csv', header=None)
phone_numbers = phone_df.iloc[:, 0] 
phone_numbers
```

```python
def get_phone_info(phone_number):
    inputs = {'PhoneNumber': phone_number, 'TestType': mTestType, 'LicenseKey': mLicenseKey}
    try:
        result = requests.get(primaryURL, params=inputs)
        outputs = xmltodict.parse(result.content)
        return outputs
    except Exception as e:
        print(f"API request failed for {phone_number}: {e}")
        return None

# Initialize variables
mTestType = "FULL"
mLicenseKey = "WSXX-XXXX-XXXX"
primaryURL = 'https://trial.serviceobjects.com/GPPL2/api.svc/GetPhoneInfo?'


def flatten_dict(d, parent_key='', sep='_'):
    items = []
    for k, v in d.items():
        new_key = f'{parent_key}{sep}{k}' if parent_key else k
        if isinstance(v, dict):
            items.extend(flatten_dict(v, new_key, sep=sep).items())
        else:
            items.append((new_key, v))
    return dict(items)
```

```python
master_df = pd.DataFrame()
for mPhoneNumber in phone_numbers:
    phone_info = get_phone_info(mPhoneNumber)
    with open('phone_info_results1.txt', 'a') as file:
        file.write(f'{mPhoneNumber, phone_info}\n')
    if phone_info:
        flat_phone_info = flatten_dict(phone_info)
        flat_phone_info = {'Phone': mPhoneNumber, **flat_phone_info}


        flat_phone_info_new = {k.replace('PhoneInfoResponse_', ''): v for k, v in flat_phone_info.items()}
        flat_phone_info_new1 = {key.replace('PhoneInfo_', '') if 'PhoneInfo_' in key else key: value for key, value in flat_phone_info_new.items()}

        xmlns_value = flat_phone_info_new1.pop('@xmlns', None)
        xmlns_i_value = flat_phone_info_new1.pop('@xmlns:i', None)
        if xmlns_value:
            flat_phone_info_new1['@xmlns'] = xmlns_value
        if xmlns_i_value:
            flat_phone_info_new1['@xmlns:i'] = xmlns_i_value    

        # phone_info_list.append(flat_phone_info_new1)


        flat_phone_info_new2 = pd.DataFrame([flat_phone_info_new1])
        master_df = pd.concat([master_df, flat_phone_info_new2], ignore_index=True)


master_df.to_csv('6.csv', header=True, index=False)
master_df
```
